## №2. Define a cube through its vertices. Scale the cube. Write down the corresponding matrix

```python
vertices = np.array([[0, 0, 0], [5, 0, 0], [5, 5, 0], [0, 5, 0],  # cube points
                     [0, 0, 5], [5, 0, 5], [5, 5, 5], [0, 5, 5]])

edges = [[0, 1], [1, 2], [2, 3], [3, 0],  # connect points to make edges
         [4, 5], [5, 6], [6, 7], [7, 4],
         [0, 4], [1, 5], [2, 6], [3, 7]]

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

for edge in edges:
    ax.plot([vertices[edge[0]][0], vertices[edge[1]][0]],
            [vertices[edge[0]][1], vertices[edge[1]][1]],
            [vertices[edge[0]][2], vertices[edge[1]][2]],
            'c-')

ax.set_xlim(-1, 6)
ax.set_ylim(-1, 6)
ax.set_zlim(-1, 6)

ax.set_box_aspect((1, 1, 1)) # do i need this

plt.show()
```

```python
scaling_factor1 = 1/2
scaling_factor2 = 2

scaling_matrix1 = np.array([[scaling_factor1, 0, 0, 0],
                            [0, scaling_factor1, 0, 0],
                            [0, 0, scaling_factor1, 0],
                            [0, 0,               0, 1]])

scaling_matrix2 = np.array([[scaling_factor2, 0, 0, 0],
                            [0, scaling_factor2, 0, 0],
                            [0, 0, scaling_factor2, 0],
                            [0, 0,               0, 1]])


reduced_vertices = np.dot(np.concatenate((vertices, np.ones((8, 1))), axis=1), scaling_matrix1.T)[:, :3]
enlarged_vertices = np.dot(np.concatenate((vertices, np.ones((8, 1))), axis=1), scaling_matrix2.T)[:, :3]

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

for edge in edges:
    ax.plot([reduced_vertices[edge[0]][0], reduced_vertices[edge[1]][0]],
            [reduced_vertices[edge[0]][1], reduced_vertices[edge[1]][1]],
            [reduced_vertices[edge[0]][2], reduced_vertices[edge[1]][2]],
            'r-')

for edge in edges:
    ax.plot([enlarged_vertices[edge[0]][0], enlarged_vertices[edge[1]][0]],
            [enlarged_vertices[edge[0]][1], enlarged_vertices[edge[1]][1]],
            [enlarged_vertices[edge[0]][2], enlarged_vertices[edge[1]][2]],
            'b-')
    
for edge in edges:
    ax.plot([vertices[edge[0]][0], vertices[edge[1]][0]],
            [vertices[edge[0]][1], vertices[edge[1]][1]],
            [vertices[edge[0]][2], vertices[edge[1]][2]],
            'c--')
    
combined_vertices = np.concatenate((reduced_vertices, enlarged_vertices), axis=0)
ax.set_xlim(-max(combined_vertices[:, 0]) - 1, max(combined_vertices[:, 0]) + 1)
ax.set_ylim(-max(combined_vertices[:, 1]) - 1, max(combined_vertices[:, 1]) + 1)
ax.set_zlim(-max(combined_vertices[:, 2]) - 1, max(combined_vertices[:, 2]) + 1)

ax.set_xlabel('X', fontsize=12)
ax.set_ylabel('Y', fontsize=12)
ax.set_zlabel('Z', fontsize=12)
ax.set_box_aspect((1, 1, 1))

plt.title('Scaled Cubes', fontsize=16)
plt.legend()

plt.show()
```
