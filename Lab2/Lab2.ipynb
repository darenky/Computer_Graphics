Побудувати матрицю повороту на кут phi навколо прямої 𝐿, що проходить через точку А(𝑎; 𝑏; 𝑐) та має направляючий вектор (𝑙; 𝑚; 𝑛). Вирішення даної задачі розбивається на кілька кроків. 1. Будуємо матрицю переміщення на вектор (−𝑎; −𝑏; −𝑐).У результаті цього переміщення ми отримуємо те, що пряма 𝐿 проходить через початок координат.
2. Суміщаємо вісь 𝑍 з прямою 𝐿 двома поворотами навколо осі 𝑋 та осі 𝑌. Перший поворот – поворот навколо осі абсцис на кут -psi. Цей кут необхідно вирахувати. Щоб знайти цей кут, розглянемо ортогональну проекцію 𝐿' заданої прямої 𝐿 на площину 𝑋 = 0 (площина 𝑌𝑂𝑍). Другий поворот – поворот навколо осі ординат на кут theta,
3. Поворот навколо прямої 𝐿 на заданий кут phi
4. Тепер виконуємо зворотні дії в порядку вкладеності. Тобто спочатку виконуємо поворот навколо осі ординат на кут -theta
5. Потім виконуємо поворот навколо осі абсцис на кут -psi
6. І останнім робимо переміщення на вектор (𝑎; 𝑏; 𝑐)
