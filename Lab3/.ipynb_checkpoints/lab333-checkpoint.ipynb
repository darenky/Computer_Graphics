## Task:

3. do 1-3 for angle θmin ≤ θ ≤ θmax. Зробити відображення у вигляді

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def test_tur0(axiom, rule, max_iter):
    for i in range(max_iter):
        new_axiom = []
        for j in range(len(axiom)):
            if axiom[j] == 'F':
                new_axiom += rule
            else:
                new_axiom += axiom[j]
        axiom = ''.join(new_axiom)
    return axiom


def plot_axiom0(axiom, fi, dfi, l):
    n = len(axiom)
    x = np.zeros(n+1)
    y = np.zeros(n+1)

    for i in range(n):
        x[i+1] = x[i]
        y[i+1] = y[i]
        if axiom[i] == 'F':
            x[i+1] = x[i+1] + l*np.cos(fi)
            y[i+1] = y[i+1] + l*np.sin(fi)
        elif axiom[i] == '+':
            fi = fi + dfi
        else:
            fi = fi - dfi
    plt.plot(x, y, 'r')
    plt.show()
```

```python
def test_tur1(axiom, ruleX, ruleY, max_iter):
    for i in range(max_iter):
        new_axiom = []
        for j in range(len(axiom)):
            if axiom[j] == 'X':
                new_axiom += ruleX
            elif axiom[j] == 'Y':
                new_axiom += ruleY
            else:
                new_axiom += axiom[j]
        axiom = ''.join(new_axiom)
    return axiom


def plot_axiom1(axiom, fi, dfi, l):
    n = len(axiom)
    x = np.zeros(n+1)
    y = np.zeros(n+1)

    for i in range(n):
        x[i+1] = x[i]
        y[i+1] = y[i]
        if axiom[i] == 'F':
            x[i+1] = x[i+1] + l*np.cos(fi)
            y[i+1] = y[i+1] + l*np.sin(fi)
        elif axiom[i] == '+':
            fi = fi + dfi
        else:
            fi = fi - dfi
    plt.plot(x, y, 'r')
    plt.show()
```

```python
def test_tur(axiom, rules, max_iter):
    for _ in range(max_iter):
        new_axiom = []
        for j in axiom:
            new_axiom.append(rules.get(j, j))  # Use get for rule lookup
        axiom = ''.join(new_axiom)
    return axiom

def plot_axiom(axiom, fi, dfi, l):
    n = len(axiom)
    x = np.zeros(n+1)
    y = np.zeros(n+1)
    for i in range(n):
        x[i+1] = x[i]
        y[i+1] = y[i]
        if axiom[i] == 'F':
            x[i+1] = x[i+1] + l*np.cos(fi)
            y[i+1] = y[i+1] + l*np.sin(fi)
        elif axiom[i] == '+':
            fi += dfi
        elif axiom[i] == '-':
            fi -= dfi
    plt.plot(x, y, 'r')
    plt.show()


fi = 0
max_iter = 5
l = 1
```

```python
rules1 = {'F': 'F-F++F-F'}
axiom1 = 'F++F++F'
dfi1 = np.pi/3
result1 = test_tur(axiom1, rules1, max_iter)
plot_axiom(result1, fi, dfi1, l)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def test_tur(axiom, ruleX=None, ruleY=None, max_iter=10):
    for _ in range(max_iter):
        new_axiom = []
        for char in axiom:
            if char == 'X' and ruleX:
                new_axiom.extend(ruleX)
            elif char == 'Y' and ruleY:
                new_axiom.extend(ruleY)
            else:
                new_axiom.append(char)
        axiom = ''.join(new_axiom)
    return axiom

def plot_axiom(axiom, fi=0, dfi=0, l=1):
    n = len(axiom)
    x = np.zeros(n+1)
    y = np.zeros(n+1)
    for i in range(n):
        if axiom[i] == 'F':
            x[i+1] = x[i] + l * np.cos(fi)
            y[i+1] = y[i] + l * np.sin(fi)
        elif axiom[i] == '+':
            fi += dfi
        elif axiom[i] == '-':
            fi -= dfi
    plt.plot(x, y, 'r')
    plt.show()

# Case 1
axiom1 = 'F++F++F'
rule1 = 'F-F++F-F'
dfi1 = np.pi/3
result1 = test_tur(axiom1, ruleX=rule1, max_iter=10)
plot_axiom(result1, fi=0, dfi=dfi1, l=2)

# Case 2
axiom2 = 'F+XF+F+XF'
ruleX2 = 'XF-F+F-XF+F+XF-F+F-X'
dfi2 = np.pi/2
result2 = test_tur(axiom2, ruleX=ruleX2, max_iter=10)
plot_axiom(result2, fi=0, dfi=dfi2, l=2)

# Case 3
axiom3 = 'YF'
ruleX3 = 'YF+XF+Y'
ruleY3 = 'XF-YF-X'
dfi3 = np.pi/3
result3 = test_tur(axiom3, ruleX=ruleX3, ruleY=ruleY3, max_iter=10)
plot_axiom(result3, fi=0, dfi=dfi3, l=2)
```
