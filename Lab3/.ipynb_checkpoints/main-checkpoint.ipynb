# LAB3. Creating fractals using L-systems

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def test_tur(axiom, rules, max_iter):   
    for _ in range(max_iter):
        new_axiom = []
        for j in axiom:
            new_axiom.append(rules.get(j, j)) 
        axiom = ''.join(new_axiom)
    return axiom

def plot_axiom(axiom, fi, dfi, l): 
    n = len(axiom)
    x = np.zeros(n+1)
    y = np.zeros(n+1)
    for i in range(n):
        x[i+1] = x[i]
        y[i+1] = y[i]
        if axiom[i] == 'F':
            x[i+1] = x[i+1] + l*np.cos(fi)
            y[i+1] = y[i+1] + l*np.sin(fi)
        elif axiom[i] == '+':
            fi += dfi
        elif axiom[i] == '-':
            fi -= dfi
    plt.plot(x, y, 'b')
    plt.show()
```

```python
fi = 0
max_iter = 5
max_iter10 = 10
l = 1
```

## 1)  Koch Snowflake (Класична сніжинка Коха)

```python
axiom1 = 'F++F++F'
rules1 = {'F': 'F-F++F-F'}
dfi1 = np.pi/3

result1 = test_tur(axiom1, rules1, max_iter)
plot_axiom(result1, fi, dfi1, l)
```

```python
dfi11 = np.pi/2
result11 = test_tur(axiom1, rules1, max_iter)
plot_axiom(result11, fi, dfi11, l)
```
